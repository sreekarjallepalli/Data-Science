```python
import numpy as np

original_array = np.arange(15, 31)
print("Original 1D array (values from 15 to 30):")
print(original_array)
print(f"Number of elements: {original_array.size}")

reshaped_array = original_array.reshape((4, 4))
print("\nReshaped 2D array (4 rows, 4 columns):")
print(reshaped_array)
print(f"Shape of the reshaped array: {reshaped_array.shape}")

reshaped_array_2 = original_array.reshape((2, 8))
print("\nReshaped 2D array (2 rows, 8 columns):")
print(reshaped_array_2)
print(f"Shape of the reshaped array: {reshaped_array_2.shape}")
```
