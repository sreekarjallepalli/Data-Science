```python
import pandas as pd

data = {
    'empid': [101, 102, 103, 104, 105],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'designation': ['Manager', 'Engineer', 'Analyst', 'Engineer', 'Designer'],
    'years of experience': [10, 5, 3, 7, 2],
    'age': [35, 28, 25, 30, 24],
    'salary': [90000, 70000, 60000, 75000, 55000]
}

employee_df = pd.DataFrame(data)

display(employee_df)
```

```python
employee_df = employee_df[employee_df['salary'] >= 10000]

display(employee_df)
```

```python
employee_df['designation'].fillna('Unknown', inplace=True)

display(employee_df)
```

```python
display(employee_df)
employee_df['age'] = employee_df['age'] + 3
```
